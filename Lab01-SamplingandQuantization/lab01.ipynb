1)	Utilizando imagem relogio.tif e faça uma reamostragem de seus pixels para 300, 150 e 72 dpi. Sabe-se que a imagem original possui 1250 dpi. Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
# type your code here
# Python code to read image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

# convertendo para 300dpi

linhas = img.shape[0]
colunas = img.shape[1]

dpi_original = 1250
dpi_novo = 300

nova_linhas = int(linhas * (dpi_novo / dpi_original))
nova_colunas = int(colunas * (dpi_novo / dpi_original))

nova_imagem = np.zeros((nova_linhas, nova_colunas), dtype=np.uint8)
for i in range(nova_linhas):
    for j in range(nova_colunas):
        x = int(i * (dpi_original / dpi_novo))
        y = int(j * (dpi_original / dpi_novo))
        x = min(x, linhas - 1)
        y = min(y, colunas - 1)
        nova_imagem[i, j] = img[x, y]

print('Tamanho da nova imagem : ',nova_imagem.shape)
plt.imshow(nova_imagem, cmap='gray', vmin=0, vmax=255)
plt.show()
```

2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
imgskull = cv2.imread("ctskull-256.tif", cv2.IMREAD_GRAYSCALE)


bits =   2 #1 , 2 , 3 , 4 , 5, 6

nivel = 2 ** bits

passo = 256 / nivel 

linhas = imgskull.shape[0]
colunas = imgskull.shape[1]

nova_imagem = np.zeros((linhas, colunas), dtype=np.uint8)
for i in range(linhas):
    for j in range(colunas):
        pixel = imgskull[i, j]
        novo_pixel = int(pixel / passo) * passo
        nova_imagem[i, j] = novo_pixel  

print("Bits:", bits)
print("Níveis:", nivel)

plt.imshow(nova_imagem, cmap='gray', vmin=0, vmax=255)
plt.title(f"{bits} bits")
plt.show()
```
