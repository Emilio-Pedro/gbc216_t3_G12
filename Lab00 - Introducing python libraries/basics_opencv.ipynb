```python
# https://www.geeksforgeeks.org/reading-image-opencv-using-python/
# Python code to read image
import cv2
import matplotlib.pyplot as plt
 
# To read image from disk, we use
# cv2.imread function, in below method,
#img = cv2.imread("foz.jpg", cv2.IMREAD_COLOR)
img = cv2.imread("beth.jpg", cv2.IMREAD_COLOR)

plt.imshow(img);
```

```python

# cv2 uses BGR color format and matplotlib uses RGB color format. To convert BGR to RGB

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB);
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img);
plt.subplot(2, 2, 2)
plt.imshow(imgRGB);
```
