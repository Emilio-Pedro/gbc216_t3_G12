```python
import numpy as np
A = np.matrix(  [ [4,3 ,2 ,1 ], [8,7,6,5],[ 12,11,10,9]  ] )   
A
```

```python
#a variável A aponta para uma matriz de duas linhas e duas colunas
type(A)
```

```python
#acessando um elemento da matriz
print("Matrix")
print(A)
print("Acessos")
print(A[0, 2])#acessa o elemento na linha 0 e na coluna 1
print(A[2]) #acessa a linha 1 da matriz
print(A[2,:]) #acessa a linha 1 da matriz. Dois pontos significa "todas" - se na primeira dimensão, todas as linhas
print(A[:, 1])  #acessa toda a coluna 0 da matriz. Dois pontos significa "todas" - se na primeira dimensão, todas as linhas

```

```python
#acessando um elemento da matriz com : (python slice) ==> inicio:fim:passo
# doc completa: https://docs.python.org/3/library/stdtypes.html#typesseq
# se início é omitido, fica com a coluna 0. Ex: :1:1
# se fim é omitido, vai até o final da dimensão. Ex 1::1
# se o passo é omitido, é assumido valor 1. Ex: 1:2
print("Matrix")
print(A)
print("Acesso [0:2,1]")
print(A[0:1, 0])#acessa o elemento na linha 0 até a linha 1 (ou seja, 2-1 = 1) e na coluna 1
print("Acessos [0:2,1:4:2]")
print(A[0:1, 0:3:1])#acessa o elemento na linha 0 até a linha 1 (ou seja, 2-1 = 1) e iniciando na coluna 1, pulando uma coluna  e indo até a colune 4-1=3
print("Mostrando a matriz com as linhas reversas")
print(A[::-1,:]) 

```

```python
# Lembre que python é orientado a objetos. Assim uma simples atribuições entre duas matrizes faz com que abas apontem para o mesmo objeto
B=A
print(A)
print(B)

print("Mudando o valor de A - veja que B foi modificado também")
A[0,0] = -20
print(A)
print(B)

print("Mudando o valor de B - veja que A foi modificado também")
B[0,0] = 40
print(A)
print(B)

```

```python
# podemos criar uma matriz copiando de uma outra matriz,
C = np.matrix(A)
print(A)
print(C)
print("Mudando o valor de A - veja que C não será afetada")
A[0,0] = 60
print(A)
print(C)


print("Mudando o valor de C - veja que A não será afetada")
C[0,0] = 40
print(A)
print(C)
```

```python
#podemos operar com blocos de matrizes
M = np.matrix(C)
print("Matriz M: ")
print(M)
print("Matriz N: ")
print(N)
#M[ 0:3,  2:4  ] = N
M[ 0:3,  0:3  ] = N
print("Matriz M após a atribuição em bloco: da linha zero até a linha 3, da coluna 2 até a 4 ")
print(M)
```

```python
#podemos fazer operações diversas com matrizes
#criando uma matriz identidade (coluna com diagonal igual a 1)
B = np.matrix(   np.eye(3,3)    )
print(B)
print(np.matrix(   np.eye(8,8)    ))

```

```python
A = np.matrix([[1 ,2], [3 ,4]])
print(A)
print(B)
#multiplicando a matriz A por um escalar
A = 4*A
A

```

```python
#somando matrizes
A = np.matrix(np.eye(3,3))
B = np.matrix(   [[1,1,1],[2,2,2],[3,3,3]]    )
print(A)
print(B)
C = A + B
C
```

```python
#multiplicando matrizes

print(A)
print(B)

D = A * B
D
```

```python
#podemos calcular potencia de matrizes

print(A)
print(B)

E = A**3 #equivale a E = A * A (multiplica a matriz por ela mesma)
E
```

```python
#podemos calcular a inversa de uma matriz

print(A)
Ainversa = A**-1
print(Ainversa)
print(A * Ainversa)
```

```python
#podemos calcular a matriz transposta
A
print(A)
print(np.transpose(A))
```

```python
#a função trace que calcula o traço da matriz (a soma dos elementos da diagonal principal)

print(B)
np.trace(B)
```

```python
#podemos concatenar matrizes
#por coluna: hstack

print(B)
print(C)
np.hstack( (B, C) )
```

```python
#concatenando por linhas (vertical): vstack

print(A)
print(B)
np.vstack(  (A,B)   )
```
